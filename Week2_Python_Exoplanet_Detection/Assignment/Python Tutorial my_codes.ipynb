```python
from astropy.io import fits
```

```python
file_dir = "data/"
file_name = "pdi_pi_collapsed.fits"
img = fits.getdata(file_dir + file_name,ext=1)
```

```python
# figur olusturma
fig=plt.figure(figsize=(8,8)) # 8*8 boyutlarında bir figur

image=plt.imshow(img,origin="lower",vmin=-0.001,vmax=0.01)
cbar=fig.colorbar(image,shrink=.80)
cbar.set_label(('Surface Brightness (arb. units)'),rotation=90,fontsize=10)
plt.xlim([60,140])
plt.ylim([60,140])
plt.xlabel("x ekseni",fontsize=12)
plt.ylabel("y ekseni",fontsize=12);

plt.savefig("2. graph.png")
```

```python
import numpy
data = numpy.loadtxt(fname='inflammation-01.csv', delimiter=',')
```

```python
ave_inflammation = numpy.mean(data, axis=0)
ave_plot = matplotlib.pyplot.plot(ave_inflammation)

matplotlib.pyplot.savefig("matplotlib2.png")
matplotlib.pyplot.show()
```

```python
import numpy
import matplotlib.pyplot

data = numpy.loadtxt(fname='inflammation-01.csv', delimiter=',')

fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.set_ylabel('average')
axes1.plot(numpy.mean(data, axis=0))

axes2.set_ylabel('max')
axes2.plot(numpy.amax(data, axis=0))

axes3.set_ylabel('min')
axes3.plot(numpy.amin(data, axis=0))

fig.tight_layout()

matplotlib.pyplot.savefig('inflammation.png')
matplotlib.pyplot.show()
```
